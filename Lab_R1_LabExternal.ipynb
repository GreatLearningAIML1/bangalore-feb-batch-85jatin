This is perhaps the best known database to be found in the pattern recognition literature. The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant.  One class is linearly separable from the other 2; the latter are NOT linearly separable from each other.

Attribute Information:
   1. sepal length in cm
   2. sepal width in cm
   3. petal length in cm

### Find out the mean and variance for each column but for class column 

```python
# Using describe function to get mean and stdev for all columns
stats_1= Iris.describe()
stats_2 = stats_1.drop("index_col", axis=1)
stats_3 = stats_2.iloc[1:3,:]
# Using transpose for variance calculation (square of stdev)
stats = stats_3.T
stats["var"] = stats["std"] * stats["std"]
stats = stats.drop("std", axis=1)
stats
```

### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
# Using describe function to store min max for all columns
range_1 = Iris.describe()
range_2 = range_1.drop("index_col", axis = 1)
range_3 = range_2.T
range_f = range_3[["min","max"]]
# Calculating range for all columns
range_f["range"] = range_f["max"] - range_f["min"]
range_f
```
