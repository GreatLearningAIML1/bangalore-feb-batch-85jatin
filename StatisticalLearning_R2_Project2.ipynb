# Q2. Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs? (10 points)
# Q3. What conclusion does the test (p-value) lead to? (2.5 points)

```python
from scipy.stats import ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro, wilcoxon
from statsmodels.stats.power import ttest_power
from scipy import stats

# Separating the data into 2 groups
pre = df['Old Scheme (in thousands)']
post = df['New Scheme (in thousands)']
```

```python
# For checking equality of variance between groups
# Null Hypothesis: Variances are equal
levene(pre,post)
print("Since p-value > 0.05, null hypothesis cannot be rejected - Variances of two groups are equal")
```

```python
# For checking Normality distribution of each distribution
# Null Hypothesis: Distribution is Normal
print("Shapiro results for 'pre' group:")
shapiro(pre)
print("Shapiro results for 'post' group:")
shapiro(post)
print("Since p-value > 0.05 for both the groups, null hypothesis cannot be rejected - Distribution is normal for both groups")
```

# Answer Q2:

```python
# Selecting paired t-test as we doing two measurments on the same experimental unit
# same set of 30 salesmen before and after change of incentive scheme
t_statistic, p_value = ttest_1samp(post - pre, 0)
print ("paired t-test p-value=", p_value)
```

# Q4.     Suppose it has been calculated that in order for Titan to break even, the average output must increase by £5000 in the scheme compared to the old scheme. If this figure is alternative hypothesis, what is: a) The probability of a type 1 error? (2.5 points) b) What is the p- value of the hypothesis test if we test for a difference of Dollars 5000? (5 points) c) Power of the test (5 points 

```python
# Calculating difference in output for each sales person
df['increase'] = df["New Scheme (in thousands)"] - df["Old Scheme (in thousands)"]
df.describe()
increase = df['increase']
```

```python
# For checking Normality distribution of distribution
# Null Hypothesis: Distribution is Normal
print("Shapiro results for 'increase':")
shapiro(increase)
print("Since p-value > 0.05 for increase, null hypothesis cannot be rejected - Distribution is normal for increase in output")
```

# Answer Q4, a): The probability of a type 1 error? (2.5 points)

```python
print("The level of significance/alpha is set at 0.05 for the one-tailed tests below, hence the probability of a type 1 error (false rejection of null hypothesis) is 0.025 (2.5 percent)")
```

```python
# Running one sample t-test for increase in output
# null hypothesis: average increase = 5 ('in thousands')
# alternative hypothesis: one-tailed, less than 5 ('in thousands')
m = 5
results = stats.ttest_1samp(increase, m)
alpha = 0.05
if (results[0] < 0) & (results[1]/2 < alpha):
    print("reject null hypothesis, mean is less than {}".format(m))
else:
    print("accept null hypothesis")
results
```

# Answer Q4, b): What is the p- value of the hypothesis test if we test for a difference of Dollars 5000? (5 points)

```python
pval = results[1]/2
print("p- value of the hypothesis test if we test for a difference of Dollars 5000: {}".format(pval))
```

```python
# Calculating Power of Test
# Compute the difference in Means between Population and Sample and divide it by Standard Deviation
# Set the number of Observations
# Set the alpha value and alternative values 'two-sided' , 'larger' , 'smaller'
val = (np.mean(increase) - 5) / np.std(increase)
powert = ttest_power(val, nobs=30, alpha=0.05, alternative='smaller')
```

# Answer Q4, c): Power of the test (5 points)

```python
print("Power of the test: {}".format(powert))
```
