```python
import pandas as pd
import numpy as np
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# read the data
df = pd.read_csv("Laliga.csv")
df.info()
```

```python
# correcting column names
df.columns = df.iloc[0]
df = df.reindex(df.index.drop(0))
```

# 1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)

```python
# Generating numeric value out of Debut and Since/LastApp 
# (example - 1931-32 will be converted to 1931 for numeric calculations)
df['Debut_year'] = df['Debut'].str.split('-').str.get(0)
df['LastApp_year'] = df['Since/LastApp'].str.split('-').str.get(0)
# replacing dashes in absolute numeric fields
cols = ['Seasons', 'Points', 'GamesPlayed', 'GamesWon', 'GamesDrawn', 'GamesLost', 'GoalsFor', 'GoalsAgainst', 
       'Champion', 'Runner-up', 'Third', 'Fourth', 'Fifth', 'Sixth', 'T', 'BestPosition', 'Debut_year', 'LastApp_year']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
# replacing NANs with zero as suggested
df = df.fillna(0)
```

```python
df.info()
```

# 2. Print all the teams which have started playing between 1930-1980. (5 points)

```python
# Filtering teams which started playing between 1930-1980, both years inclusive
df_between1930_1990 = df.loc[(df["Debut_year"] > 1929) & (df["Debut_year"] < 1981)]
```

# 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

```python
# function definition
def Goal_diff_count(GoalsFor,GoalsAgainst):
    diff = GoalsFor - GoalsAgainst
    return diff
```

```python
# calculating goal difference using function defined above
df['Goal_diff'] = Goal_diff_count(df['GoalsFor'], df['GoalsAgainst'])
team_maxGD = df.sort_values('Goal_diff',ascending = False).head(1)
team_minGD = df.sort_values('Goal_diff',ascending = True).head(1)
```

```python
print('Team which has maximum goal difference:')
team_maxGD.iloc[:, [1,2,8,9,22]]
print('Team which has minimum goal difference:')
team_minGD.iloc[:, [1,2,8,9,22]]
```

# 6. Group teams based on their “Best position” and print the sum of their points for all positions (10 points)

```python
# Group by best position
best_position = df.groupby("BestPosition")["Points"].agg("sum")

print("Sum of grouped teams' points for all positions:")
pd.DataFrame({'BestPosition':best_position.index, 'Points':best_position.values})
```
